```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import math
import numpy as np
import networkx as nx
from operator import itemgetter
```

```python
path = "../Dataset Super Users/edges.csv"
path_all = "../Dataset Super Users/questions_2020_2022.csv"
path_nodes = "../Dataset Super Users/nodes.csv"
```

```python
df = pd.read_csv(path)
```

```python
df
```

```python
G = nx.read_edgelist(path, comments = "S", delimiter = ",", data=(("Count", int),))
```

```python
degree_dict = dict(G.degree(G.nodes()))
nx.set_node_attributes(G, degree_dict, 'degree')
sorted_degree = sorted(degree_dict.items(), key = itemgetter(1), reverse = True)
print(" ---- Top 150 nodes by degree ----")
for d in sorted_degree[:150]:
    print("\t" + str(d))
```

```python
nodes = []
others = []
for i in range(len(sorted_degree)):
    if i<10002:
        nodes.append(sorted_degree[i][0])
    else:
        others.append(sorted_degree[i][0])
```

```python
new_df = df[(df["Source"].isin(nodes)) & (df["Target"].isin(nodes))] 
```
